```python
# Calculate the average 'lowest_yearly_earnings' for each category
category_averages = df.groupby('category')['lowest_yearly_earnings'].mean()

# Function to apply the condition for earnings less than 100
def adjust_earnings(row):
    if row['lowest_yearly_earnings'] < 10000:
        return category_averages[row['category']]
    else:
        return row['lowest_yearly_earnings']

# Apply the function to adjust the 'lowest_yearly_earnings' column
df['lowest_yearly_earnings'] = df.apply(adjust_earnings, axis=1)

# Check the updated 'lowest_yearly_earnings' for each category
df.groupby('category')['lowest_yearly_earnings'].min()
```

```python
# Calculate the average 'highest_yearly_earnings' for each category
category_high_averages = df.groupby('category')['highest_yearly_earnings'].mean()

# Function to apply the condition for earnings less than 100
def adjust_high_earnings(row):
    if row['highest_yearly_earnings'] < 10000:
        return category_high_averages[row['category']]
    else:
        return row['highest_yearly_earnings']

# Apply the function to adjust the 'highest_yearly_earnings' column
df['highest_yearly_earnings'] = df.apply(adjust_high_earnings, axis=1)

# Check the updated 'highest_yearly_earnings' for each category
df.groupby('category')['highest_yearly_earnings'].min()
```

```python
# Calculate the category averages for the specified columns with a different variable name
category_mean_earnings = df.groupby('category')[['lowest_monthly_earnings', 'highest_yearly_earnings']].mean()

# Define a function to replace values less than 1000 with category averages, using different variable names
def replace_with_avg(row, column_name):
    if row[column_name] < 1000:
        return category_mean_earnings.loc[row['category'], column_name]
    return row[column_name]

# Apply the function to the necessary columns with modified names
df['lowest_monthly_earnings'] = df.apply(replace_with_avg, column_name='lowest_monthly_earnings', axis=1)
df['highest_yearly_earnings'] = df.apply(replace_with_avg, column_name='highest_yearly_earnings', axis=1)

# Show the modified data to check the changes
df[['Youtuber', 'category', 'lowest_monthly_earnings', 'highest_yearly_earnings']].head()
```
